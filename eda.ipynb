# Basic Dataset Info

### Key Findings:

* Rows where 'fipscode' is null will be dropped
* 9020 county rows for 'Uninsured' and 'Children in poverty' measures are missing FIPS codes due to a data sourcing issue.
* 561 state/national aggregate rows will be dropped as out of scope.

```python
import pandas as pd

df = pd.read_csv('./data/County_Health_Rankings.csv')

# Display basic information
print(f"Dataset Shape: {df.shape}")
df.head(10)
```

```python
print("Dataset Overview:")
print("=" * 50)
print(f"Total samples: {len(df)}\n")
print(df.info())
```

```python
measure_names_df = pd.DataFrame({'Measure Names': df['Measure name'].unique()})
print("Unique Measure Names:")
display(measure_names_df)
```

```python
year_span_df = pd.DataFrame({'Year spans': df['Year span'].unique()})
print("Unique Year Spans:")
display(year_span_df)
```

## Handling Missing Values

```python
# Checking for missing values
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100
missing_df = pd.DataFrame({'Missing Values': missing_values, 
                           'Percentage (%)': missing_percentage})
missing_df = missing_df[missing_df['Missing Values'] > 0].sort_values('Percentage (%)', ascending=False)

print("Missing Values Analysis:")
display(missing_df)
```
